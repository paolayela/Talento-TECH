```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

#Cargamos la imagen
vaca=cv2.imread('VACA.png')

#Definimos el filtro
filtro=np.array([[0,-1,0],
                 [-1,5,-1],
                 [0,-1,0]])

#Aplicamos el filtro
vaca_filtrada=cv2.filter2D(vaca,-1,filtro)

#mostramos la imagen original y la imagen convolucionada
cv2_imshow(vaca)
cv2_imshow(vaca_filtrada)

#esperamos a que se cierre la ventana
cv2.waitKey(0)
cv2.destroyAllWindows()
```
