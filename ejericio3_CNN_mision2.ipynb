```python
from google.colab.patches import cv2_imshow
!pip install openCV-python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#cargar la imagen
pajaro=cv2.imread('pajaro1.jpg')

#mostrar la imagen en una ventana
cv2_imshow(pajaro)

#esperar a que se cierre la ventana
cv2.waitKey(0)

#destruir la ventana
cv2.destroyAllWindows()
```

```python
#optener las dimensiones de la imagen
heigth,width=pajaro.shape[:2]

#calcular el centro de la imagen
centro=(width/2,heigth/2)

#definir la matriz de rotacion
rotacion=45
M=cv2.getRotationMatrix2D(centro,rotacion,1)

#aplicar la rotacion a la imagen
pajaro_rotacionado=cv2.warpAffine(pajaro,M,(width,heigth))

#mostrar la imagen en una ventana
cv2_imshow(pajaro_rotacionado)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Traslacion

pajaro=cv2.imread('pajaro1.jpg')

#definir la matriz de traslacion
tx,ty=100,50
M=np.float32([[1,0,tx],[0,1,ty]])

#Aplicar la traslacion a la imagen
pajaro_trasladado=cv2.warpAffine(pajaro,M,(pajaro.shape[1],pajaro.shape[0]))

#mostrar la imagen en una ventana
cv2_imshow(pajaro_trasladado)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#escala
pajaro=cv2.imread('pajaro1.jpg')

#definir la nueva altura y anchura
nueva_Altura,nueva_anchura=300,150

#calcular la matriz de escala
scalar=cv2.resize(pajaro,(nueva_Altura,nueva_anchura))

#mostrar la imagen en una ventana
cv2_imshow(scalar)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#suavizado

#aplicar el filtro Gaussiano para suavizar la imagen
suavizado=cv2.GaussianBlur(pajaro,(5,5),0)

#mostrar la imagen en una ventana
cv2_imshow(suavizado)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Realce

#definir el kernel para el filtro de afilado
kernel=np.array([[-1,-1,-1],
                [-1,9,-1],
                [-1,-1,-1]])

#Aplicar el filtro de afilado para realzar los detalles
realce=cv2.filter2D(pajaro,-1,kernel)

#mostrar la imagen en una ventana
cv2_imshow(realce)
cv2.waitKey(0)
cv2.destroyAllWindows
```

```python
#Deteccion de bordes

#Aplicar el operador sobel para detectar los bordes
Sobel_x=cv2.Sobel(pajaro,cv2.CV_64F,1,0,ksize=5)
sobel_y=cv2.Sobel(pajaro,cv2.CV_64F,0,1,ksize=5)

#cambiar las respuestas en magnitudes
magnitudes=cv2.magnitude(Sobel_x,sobel_y)

#Normalizar los valores para mostrar la imagen correcta
magnitudes=cv2.normalize(magnitudes,None,0,255,cv2.NORM_MINMAX,cv2.CV_8U)


#mostrar la imagen en una ventana
cv2_imshow(magnitudes)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#cargamos la imagen
rosa=cv2.imread('rosa.png')

#Miramos el alto y el ancho de la imagen
alto,ancho=rosa.shape[:2]

#definimos la matriz de rotacion
rotacion=90
M=cv2.getRotationMatrix2D((ancho/2,alto/2),rotacion,1)

#Aplicamos la rotacion a la imagen
rosa_rotada=cv2.warpAffine(rosa,M,(ancho,alto)) # Changed the shape here

# Crear una imagen grande para mostrar ambas imgenes una al lado de la otra
imagen_grande = cv2.hconcat([rosa, rosa_rotada])

#mostramos la imagen en una ventana
cv2_imshow(imagen_grande)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#cargar la imagen
rosa=cv2.imread('rosa.png')

#aplicar el filtro Gaussiano para suavizar la imagen
suavizado=cv2.GaussianBlur(rosa,(5,5),0)

#definir el kernel para el filtro de afilado
kernel=np.array([[-1,-1,-1],
                [-1,9,-1],
                [-1,-1,-1]])

#Aplicar el filtro de afilado para realzar los detalles
realce=cv2.filter2D(rosa,-1,kernel)

# Crear una imagen grande para mostrar ambas imgenes una al lado de la otra
imagen_grande = cv2.hconcat([suavizado, realce])


#mostrar la imagen en una ventana
cv2_imshow(imagen_grande)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Aplicar el operador sobel para detectar los bordes
Sobel_x=cv2.Sobel(rosa,cv2.CV_64F,1,0,ksize=5)
sobel_y=cv2.Sobel(rosa,cv2.CV_64F,0,1,ksize=5)

#cambiar las respuestas en magnitudes
magnitudes=cv2.magnitude(Sobel_x,sobel_y)

#Normalizar los valores para mostrar la imagen correcta
magnitudes=cv2.normalize(magnitudes,None,0,255,cv2.NORM_MINMAX,cv2.CV_8U)

# Crear una imagen grande para mostrar ambas imgenes una al lado de la otra
imagen_grande = cv2.hconcat([rosa, magnitudes])

#mostrar la imagen en una ventana
cv2_imshow(imagen_grande)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
